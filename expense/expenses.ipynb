## Tools Used
- Python
- Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset Overview

The dataset contains daily expense records with the following fields:

- Date — transaction date
- Category — type of expense
- Amount — money spent

This dataset simulates real-world personal finance tracking.



```python
df = pd.read_csv("expenses.csv")
df.head()
```

## Key Questions

- What is the total spending?
- Which categories drive the highest expenses?
- Are there spending spikes?
- What is the average daily spend?
- What insights can support better budgeting?

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
total = df["Amount"].sum()
print("Total Spending:", total)
```

```python
category = df.groupby("Category")["Amount"].sum()
print(category)
```

```python
category.plot(kind="bar")
plt.title("Spending by Category")
plt.show()
```

Observation: The bar chart shows that Bills account for the highest portion of spending, followed by Shopping and Food. This indicates that fixed expenses are the primary cost drivers, while discretionary spending also contributes significantly.

```python
category.plot(kind="pie", autopct="%1.1f%%")
plt.title("Expense Distribution")
plt.ylabel("")
plt.show()
```

Observation:
Bills represent the largest portion of spending, while other categories contribute smaller shares.

```python
daily = df.groupby("Date")["Amount"].sum()
daily.plot()
plt.title("Daily Spending Trend")
plt.show()
```

Observation: Spending varies across days with spikes indicating major expenses.

```python
df["Month"] = df["Date"].dt.to_period("M")

monthly = df.groupby("Month")["Amount"].sum()
print(monthly)

monthly.plot(kind="bar")
plt.title("Monthly Spending Summary")
plt.show()
```

Observation: Since the dataset covers a single month, the chart reflects total spending for that period and serves as a baseline for comparing future months.

```python
highest_day = df.loc[df["Amount"].idxmax()]
print("Highest Spending Day:")
print(highest_day)
```

Observation: The highest expense occurred due to a bill payment, indicating that fixed obligations create significant spending spikes.

```python
avg_daily = df.groupby("Date")["Amount"].sum().mean()
print("Average Daily Spending:", avg_daily)
```

Observation: The average daily spending provides a useful benchmark for tracking expenses and maintaining budget discipline.

```python
top3 = category.sort_values(ascending=False).head(3)
print("Top 3 Categories:")
print(top3)
```

Observation: Bills, Shopping, and Food are the main contributors to total expenses, suggesting these areas should be prioritized for cost management.

## Recommendations

- Monitor bill payments closely as they form the largest expense portion.
- Track discretionary spending such as shopping to reduce unnecessary costs.
- Use average daily spending as a budgeting benchmark.
- Identify high-spending days to plan better financial control.
